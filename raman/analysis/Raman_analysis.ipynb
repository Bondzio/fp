```python
files = ["mono_slit_100", "mono_slit_150","mono_slit_200","mono_slit_25", \
      "mono_slit_25_02","mono_slit_40", "mono_slit_50","mono_slit_75"]
fig, ax = plt.subplots(1,1)
colors = sns.diverging_palette(220, 20, n = len(files))
for n, filename in enumerate(files): 
    lamb = np.load("npy/"+filename+"_lamb.npy")/10
    count = np.load("npy/"+filename+"_count.npy")
    
    ax.plot(lamb,count, "-", alpha = 0.8,label = "$"+find_number(filename)+"$", color= colors[n])
    ax.scatter(lamb,count, alpha = 0.8, color= colors[n],s = 60)
ax.legend(title="width of gap/ $\mu m$", loc =2)
ax.set_ylim(0,500000)
ax.set_xlim(404,405)
ax.set_xlabel("wavelength $\lambda$ in $nm$")
ax.set_ylabel("counts")
ax.set_title("Monochromator with white light: Full spectrum")
ax.grid(True)
plt.show()
```

```python
filename = "mono_chcl3_01"
lamb1 = np.load("npy/"+filename+"_lamb.npy")/10
count1 = np.load("npy/"+filename+"_count.npy")

filename = "mono_chcl3_02"
lamb2 = np.load("npy/"+filename+"_lamb.npy")/10
count2 = np.load("npy/"+filename+"_count.npy")

fig, ax = plt.subplots(1,1)
back_mean = count1[(lamb1 > 515)*(lamb1 < 525)].mean()
back_var  = count1[(lamb1 > 515)*(lamb1 < 525)].std()
ax.plot(lamb1,(count1-back_mean)/back_var, ".", alpha = 0.2)
ax.plot(lamb1,(count1-back_mean)/back_var, "-", alpha = 0.2, color = "green")

back_mean = count2[(lamb2 > 515)*(lamb2 < 525)].mean()
back_var  = count2[(lamb2 > 515)*(lamb2 < 525)].std()
ax.plot(lamb2,(count2-back_mean)/back_var, ".", alpha = 0.4)
ax.plot(lamb2,(count2-back_mean)/back_var, "-", alpha = 0.4, color = "orange")

#ax.plot(lamb, savitzky_golay(count,15,3))
#ax.fill_between(lamb,count - np.sqrt(count),count + np.sqrt(count))
ax.set_ylim(-6,10)
#ax.set_yscale("log")
ax.set_xlim(515,550)
ax.set_xlabel("wavelength $\lambda$ in $nm$")
ax.set_ylabel("counts")
ax.set_title("Monochromator with white light: Full spectrum")
ax.grid(True)
plt.show()
```

```python
filename = "mono_chcl3_01"
lamb1 = np.load("npy/"+filename+"_lamb.npy")/10
count1 = np.load("npy/"+filename+"_count.npy")

filename = "mono_chcl3_02"
lamb2 = np.load("npy/"+filename+"_lamb.npy")/10
count2 = np.load("npy/"+filename+"_count.npy")

fig, ax = plt.subplots(1,1)


back_mean = count1[(lamb1 > 515)*(lamb1 < 525)].mean()
max_val = np.max((count1-back_mean)[(lamb1 > 538)*(lamb1 < 545)])

ax.plot(lamb1,(count1-back_mean)/max_val, ".", alpha = 0.2)
ax.plot(lamb1,(count1-back_mean)/max_val, "-", alpha = 0.2, color = "green")

back_mean = count2[(lamb2 > 515)*(lamb2 < 525)].mean()
max_val = np.max((count2-back_mean)[(lamb2 > 538)*(lamb2 < 545)])

ax.plot(lamb2,(count2-back_mean)/max_val, ".", alpha = 0.4)
ax.plot(lamb2,(count2-back_mean)/max_val, "-", alpha = 0.4, color = "orange")

#ax.plot(lamb, savitzky_golay(count,15,3))
#ax.fill_between(lamb,count - np.sqrt(count),count + np.sqrt(count))
ax.set_ylim(-0.5,1)
#ax.set_yscale("log")
ax.set_xlim(538,545)
ax.set_xlabel("wavelength $\lambda$ in $nm$")
ax.set_ylabel("counts")
ax.set_title("Monochromator with $CHCl_3$: Full spectrum")
ax.grid(True)
plt.show()
```
